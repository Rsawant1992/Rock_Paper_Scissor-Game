```python
from random import randint

option_choices = ["rock", "paper", "scissors"]

def player_input():
    
    while True:
        
        print("Type 0 for rock")
        print("Type 1 for paper")
        print("Type 2 for scissors")
    
        try:
            player = int(input("-->"))
            player_choice = option_choices[player]
            return player_choice
    
        except ValueError:
            print("You Entered a text instead of a Number")
        
        except IndexError:
            print("You Entered a Number that was out of range. Please enter between 0-2")
            
        

def Game_Round():
    
    player_choice = player_input()
    
    computer_choice = option_choices[randint(0,2)]
    print("You picked: " + player_choice)
    print("Computer picked: " + computer_choice)
    
    if player_choice == computer_choice:
        print("\tGame Drawn :)")
        
        
    elif player_choice == "rock" and computer_choice == "scissors":
        print("\tYou won! :)")
        return "player"
    
    elif player_choice == "paper" and computer_choice == "rock":
        print("\tYou won! :)")
        return "player"
        
    elif player_choice == "scissors" and computer_choice == "paper":
        print("\tYou won! :)")
        return "player"
    
    else:
        print("\tThe computer Won! :(")
        return "computer"

    

def Game():
    
    c = 0
    player_wins = 0
    print(c)
    
    for i in range(5):
        winner = Game_Round()        
        
        if winner == "player":
            print("1")
            player_wins += 1
    
        elif winner == "computer":
            print("2")
            c += 1
    
    print("The computer won: "+ str(c))
    print("You won: "+  str(player_wins))
    


Game()
    

    
```
